```python
import numpy as np
import matplotlib as mp
from matplotlib import pyplot as plt
```

Simple Montecarlo

```python
y=[]
for i in range(1,1001):
    I=0.
    n0=i
    N=n0
    r=np.random.random(n0)
    for i in range(n0):
        x=r[i]
        I+=np.sqrt(1-x**2)
    I*=4./float(n0)
    y.append(I)
x=np.arange(1,1001,1)
```

```python
fig, ax=plt.subplots()
ax.plot(x,y, 'x',color='Blue')
fig.set_size_inches(5,5)
plt.show()
```

```python
plt.xlim(0,1,.1)
x2=np.arange(0,1,.01)
plt.plot(x2,np.sqrt(1-x2**2))
```

Importance Sampling

```python
#Simplified MC
n0=10000
r=np.random.random(n0)
Itot=np.sum(r**1.5*np.exp(-r))
Itot*=1/float(n0)
#Importance Sampling MC, with importance function f(x)=e^-x
x=-np.log(r)
I2=np.sum(x**1.5)
I2*=1/float(n0)
print("Simplified Montecarlo:", Itot)
print("Importance Sampling Montecarlo:", I2)
```
